```python
import whisper
import cv2
import numpy as np
from moviepy.editor import VideoFileClip, TextClip, CompositeVideoClip
import re
import os
from nltk.tokenize import sent_tokenize
import nltk

# def create_technical_keywords():
#     return {
#         'gpu_computing': [
#             'gpu', 'cuda', 'parallel computing', 'graphics processing', 'shader',
#             'compute shader', 'opencl', 'vulkan', 'graphics pipeline', 'rendering',
#             'texture', 'buffer', 'compute unit', 'thread block', 'warp', 'kernel'
#         ],
#         'computer_architecture': [
#             'instruction set', 'isa', 'pipeline', 'branch prediction', 'cache hierarchy',
#             'memory hierarchy', 'von neumann', 'harvard architecture', 'superscalar',
#             'out of order execution', 'speculative execution', 'microarchitecture',
#             'fetch', 'decode', 'execute', 'writeback', 'forwarding', 'hazard',
#             'stall', 'microcode', 'microinstruction'
#         ],
#         'assembly_programming': [
#             'assembly', 'assembler', 'mnemonic', 'opcode', 'operand', 'risc',
#             'cisc', 'arm assembly', 'x86 assembly', 'riscv', 'risc-v', 'instruction set',
#             'register file', 'immediate value', 'addressing mode', 'branch instruction',
#             'jump instruction', 'load store', 'arithmetic instruction', 'logical instruction'
#         ],
#         'low_level': [
#             'memory', 'pointer', 'address', 'register', 'cache', 'assembly',
#             'instruction', 'binary', 'bit', 'byte', 'stack', 'heap', 'allocation',
#             'memory mapping', 'virtual memory', 'physical memory', 'page table',
#             'segmentation', 'protection ring', 'privilege level'
#         ],
#         'system_programming': [
#             'operating system', 'driver', 'interrupt', 'system call', 'process',
#             'thread', 'scheduling', 'synchronization', 'mutex', 'semaphore',
#             'context switch', 'privilege level', 'kernel mode', 'user mode'
#         ],
#         'hardware': [
#             'cpu', 'processor', 'core', 'bus', 'dma', 'pipeline', 'architecture',
#             'bandwidth', 'latency', 'throughput', 'clock', 'frequency', 'alu',
#             'control unit', 'floating point unit', 'vector unit', 'simd'
#         ],
#         'optimization': [
#             'performance', 'optimization', 'efficient', 'bottleneck', 'profiling',
#             'benchmark', 'throughput', 'latency', 'parallel', 'concurrent',
#             'vectorization', 'loop unrolling', 'instruction scheduling'
#         ]
#     }

# Download required NLTK data
nltk.download('punkt')

def transcribe_video(video_path):
    """Transcribe video using Whisper"""
    model = whisper.load_model("base")
    result = model.transcribe(video_path)
    return result["segments"]

def generate_clip_title(text):
    """Generate a meaningful title from clip content"""
    # Define topic patterns with their corresponding titles
    topic_patterns = [
        (r'gpu\s+architecture|graphics\s+processing|cuda\s+programming',
         'GPU Architecture and Programming'),
        (r'memory\s+management|cache\s+hierarchy|shared\s+memory',
         'Memory Systems'),
        (r'parallel\s+processing|thread\s+execution|concurrent',
         'Parallel Computing'),
        (r'assembly\s+language|instruction\s+set|machine\s+code',
         'Assembly Programming'),
        (r'optimization|performance|efficiency',
         'Performance Optimization'),
        (r'compiler|translation|code\s+generation',
         'Compiler Technology'),
        (r'hardware|processor|circuit',
         'Computer Hardware'),
    ]

    text_lower = text.lower()

    # Try to find a matching topic
    for pattern, title in topic_patterns:
        if re.search(pattern, text_lower):
            # Extract the specific context
            context = re.search(r'([^.!?]+[.!?])', text)
            if context:
                return f"{title}: {context.group(1).strip()}"
            return title

    # Default to first sentence if no specific topic is found
    first_sentence = re.search(r'([^.!?]+[.!?])', text)
    if first_sentence:
        return first_sentence.group(1).strip()
    return "Technical Discussion"

def find_technical_segments(segments, min_duration=45, max_duration=90, max_clips=5):
    """Find technical segments with proper duration"""
    technical_terms = {
        'gpu', 'cuda', 'memory', 'thread', 'kernel', 'assembly', 'instruction',
        'cache', 'pipeline', 'register', 'parallel', 'processor', 'compiler',
        'optimization', 'architecture', 'algorithm', 'buffer', 'compute',
        'graphics', 'hardware', 'software', 'programming', 'code', 'data'
    }

    potential_clips = []
    current_clip = None

    for segment in segments:
        text = segment["text"].lower()

        # Check if segment contains technical content
        is_technical = any(term in text for term in technical_terms)

        if is_technical:
            if current_clip is None:
                current_clip = {
                    "start": segment["start"],
                    "end": segment["end"],
                    "text": segment["text"],
                    "segments": [segment]
                }
            else:
                current_clip["end"] = segment["end"]
                current_clip["text"] += " " + segment["text"]
                current_clip["segments"].append(segment)

            # Check if current clip duration is within bounds
            duration = current_clip["end"] - current_clip["start"]
            if duration >= min_duration:
                if duration <= max_duration:
                    potential_clips.append(current_clip)
                current_clip = None

        elif current_clip is not None:
            # Add one more segment for context
            current_clip["end"] = segment["end"]
            current_clip["text"] += " " + segment["text"]
            current_clip["segments"].append(segment)

            duration = current_clip["end"] - current_clip["start"]
            if duration >= min_duration:
                if duration <= max_duration:
                    potential_clips.append(current_clip)
                current_clip = None

    # Sort clips by technical content density and take top max_clips
    clips = sorted(potential_clips,
                  key=lambda x: sum(1 for term in technical_terms if term in x["text"].lower()),
                  reverse=True)

    return clips[:max_clips]

def create_subtitle_clip(text, video_width, video_height, duration):
    """Create subtitle clip with proper formatting"""
    txt_clip = TextClip(
        text,
        font='Arial',
        fontsize=30,
        color='white',
        bg_color='black',
        size=(video_width - 20, None),
        method='caption'
    )
    txt_clip = txt_clip.set_position(('center', 'bottom'))
    txt_clip = txt_clip.set_duration(duration)
    return txt_clip

def process_clip(video_path, clip_info, output_path):
    """Process a single video clip with subtitles"""
    try:
        with VideoFileClip(video_path) as video:
            # Add small padding to avoid cutting off speech
            start_time = max(0, clip_info["start"] - 0.5)
            end_time = min(video.duration, clip_info["end"] + 0.5)

            main_clip = video.subclip(start_time, end_time)

            # Create subtitles
            subtitle_clips = []
            for segment in clip_info["segments"]:
                if segment["start"] >= start_time and segment["start"] < end_time:
                    sub_start = segment["start"] - start_time
                    sub_duration = min(segment["end"] - segment["start"],
                                     end_time - segment["start"])

                    subtitle = create_subtitle_clip(
                        segment["text"],
                        main_clip.w,
                        main_clip.h,
                        sub_duration
                    ).set_start(sub_start)

                    subtitle_clips.append(subtitle)

            # Combine video with subtitles
            final_clip = CompositeVideoClip([main_clip] + subtitle_clips)

            # Generate title from content
            title = generate_clip_title(clip_info["text"])

            # Create filename from title
            safe_title = re.sub(r'[^\w\s-]', '', title)
            safe_title = re.sub(r'[-\s]+', '_', safe_title)
            final_output_path = f"{output_path}_{safe_title}.mp4"

            # Write final clip
            final_clip.write_videofile(
                final_output_path,
                codec='libx264',
                audio_codec='aac',
                temp_audiofile="temp-audio.m4a",
                remove_temp=True,
                threads=4
            )

            return title, final_output_path

    except Exception as e:
        print(f"Error processing clip: {str(e)}")
        return None, None

def extract_technical_clips(video_path, output_prefix="clip", max_clips=5):
    """Main function to extract technical clips"""
    print("Transcribing video...")
    segments = transcribe_video(video_path)

    print("Finding technical segments...")
    technical_clips = find_technical_segments(
        segments,
        min_duration=45,  # Minimum 45 seconds
        max_duration=90,  # Maximum 90 seconds
        max_clips=max_clips  # Maximum number of clips to generate
    )

    print(f"Found {len(technical_clips)} technical segments")

    for i, clip_info in enumerate(technical_clips, 1):
        print(f"\nProcessing clip {i}/{len(technical_clips)}...")

        title, output_path = process_clip(video_path, clip_info, output_prefix)

        if title:
            print(f"Created clip: {title}")
            print(f"Duration: {clip_info['end'] - clip_info['start']:.2f} seconds")
            print(f"Saved as: {output_path}")

if __name__ == "__main__":
    video_path = "downloads/video.mp4"
    # Configure maximum number of clips (default 5)
    max_clips = 3
    extract_technical_clips(video_path, "technical_clip", max_clips)

# Created/Modified files during execution:
# - technical_clip_[Title].mp4 files
```
